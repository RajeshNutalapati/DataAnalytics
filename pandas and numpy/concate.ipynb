```python
import pandas as pd
india_weather = pd.DataFrame({
    'city': ["mumbai", "delhi", "pune"],
    'temperature':[22,33,44],
    'humidity':[30,60,30]
})

india_weather
```

```python
us_weather = pd.DataFrame({
    'city': ["NY", "chicago", "Washington"],
    'temperature':[20,30,40],
    'humidity':[75,65,35]
})
us_weather

df = pd.concat([india_weather, us_weather])
df #watch the index

df = pd.concat([india_weather, us_weather], ignore_index=True)
df
#you can use keys to check for india weather only

df = pd.concat([india_weather, us_weather], keys=["india", "us"])
df

#you can use loc for location

df.loc["india"]

#if you want to append to columns

windspeed_df = pd.DataFrame({
    "city":["mumbai", "delhi", "pune"],
    "windspeed":[2,3,4]
})
windspeed_df

df=pd.concat([india_weather, windspeed_df], sort=False)
df #observe the output of the index, we dont want that. we want to merge the data, use axis here

df=pd.concat([india_weather, windspeed_df], sort=False, axis=1)
df
```

```python
# now say if the 2 dataframes are not similar
#and then you need to concatinate. For exame, windspeed is missing mumbai and "delhi and pune" 
#are not in order. You can use index in such scenarios
```

```python
windspeed_df = pd.DataFrame({
    "city":["pune", "delhi"],
    "windspeed":[2,3]
}, index = [1,0])
windspeed_df

temperature_df = pd.DataFrame({
    'city': ["mumbai", "delhi", "pune"],
    'temperature':[22,33,44]
}, index=[0,1,2])

df = pd.concat([windspeed_df, temperature_df], sort=False, axis =1)
df

```

```python
#append series to df
s = pd.Series(["rain", "humid", "dry"], name="event")
s
```

```python
df = pd.concat([temperature_df, s], axis=1)
df
```

```python
#MERGE
df2 = pd.merge(temperature_df, windspeed_df, on="city")
df2  #one city is missing, this is intersection(inner is intersection and outer is union)
```

```python
#left join and right join
df2 = pd.merge(temperature_df, windspeed_df, on="city", how="left")
df2
```

```python
#use indicator to know from where the data came
df2 = pd.merge(temperature_df, windspeed_df, on="city", how="left", indicator=True)
df2
```

```python
# now notice another difference for repeatative colums in 2 df
windspeed_df = pd.DataFrame({
      'city': ["mumbai", "delhi", "pune"],

    "windspeed":[2,3,4],
    'temperature':[22,33,44]
})
windspeed_df

temperature_df = pd.DataFrame({
    'city': ["mumbai", "delhi", "pune"],
    'temperature':[22,33,44]
})

df3 = pd.merge(windspeed_df, temperature_df, on="city")
df3

# observe suffix, u can also supply suffix


```

```python
df3 = pd.merge(windspeed_df, temperature_df, on="city", suffixes=('_left','_right'))
df3
```
